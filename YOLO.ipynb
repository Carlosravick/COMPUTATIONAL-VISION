```python
from ultralytics import YOLO


model = YOLO('yolo11m.pt')

# Treinamento
model.train(
    data='/content/drive/MyDrive/sku/data.yaml',
    epochs=70,
    imgsz=640,
    batch=16,
    project='runs/train',  # padrÃ£o
    name='exp',            # padrÃ£o
    exist_ok=True          # sobrescreve se jÃ¡ existir
)
```

```python
from ultralytics import YOLO
import os
from PIL import Image

# Caminho da imagem original
img_path = '/content/drive/MyDrive/sku/test/skutest.jpeg'

# Caminho temporÃ¡rio da imagem redimensionada
resized_img_path = '/content/drive/MyDrive/sku/test/skutest_resized.jpeg'

# Redimensiona a imagem com PIL
img = Image.open(img_path)
img_resized = img.resize((640, 640))
img_resized.save(resized_img_path)

# Carrega o modelo treinado
model = YOLO('runs/train/exp/weights/best.pt')

# Faz a prediÃ§Ã£o na imagem redimensionada
results = model.predict(resized_img_path, imgsz=640, conf=0.50)

# Imprime quantas detecÃ§Ãµes teve
print(f'NÃºmero de objetos detectados: {len(results[0].boxes)}')

# Imprime as classes e confianÃ§as
for box in results[0].boxes:
    print(f'Classe: {int(box.cls)}, ConfianÃ§a: {box.conf.item():.2f}')

# DiretÃ³rio de salvamento
save_dir = '/content/drive/MyDrive/sku/results'
os.makedirs(save_dir, exist_ok=True)
save_path = os.path.join(save_dir, os.path.basename(resized_img_path))
results[0].save(filename=save_path)

# Mostra a imagem detectada (Jupyter)
results[0].show()
```

```python
from ultralytics import YOLO
import cv2
import os
from google.colab.patches import cv2_imshow

# Caminho da imagem original
img_path = '/content/drive/MyDrive/sku/test/Q50.jpg'

# Carrega o modelo treinado
model = YOLO('runs/train/exp/weights/best.pt')

# Faz a prediÃ§Ã£o
results = model.predict(img_path, imgsz=640, conf=0.25)

# Converte a imagem original para OpenCV (BGR)
img = cv2.imread(img_path)

# Itera sobre os boxes detectados e desenha na imagem
for box in results[0].boxes:
    x1, y1, x2, y2 = map(int, box.xyxy[0])
    cls = int(box.cls)
    conf = float(box.conf)

    # Desenha a caixa
    cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)

    # Texto com classe e confianÃ§a
    label = f'{cls} ({conf:.2f})'
    cv2.putText(img, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX,
                0.6, (0, 255, 0), 2)

# Mostra a imagem em uma janela usando cv2_imshow
cv2_imshow(img)
# cv2.waitKey(0)  # Espera atÃ© qualquer tecla ser pressionada - not needed with cv2_imshow
# cv2.destroyAllWindows() # not needed with cv2_imshow
```
